## 8. Escribe una funci贸n que devuelva la moda de una lista, es decir: el elemento que aparece m谩s veces.

`NOTE: No se debe usar count... `

```python
def mode_counter(arr):
    frecuencias = {}
    for elemento in arr:
        if elemento not in frecuencias:
            frecuencias[elemento] = 1
        if elemento in frecuencias:
            frecuencias[elemento] += 1

    return max(frecuencias, key=frecuencias.get)
```

```python
# Para testear tu funci贸n 
test_mode(mode_counter)
```

```python
def pangram(string):
    string = string.lower()
    conjunto = set(letter for letter in string if letter.isalpha())
    if len(conjunto) == 26:
        return True
    else:
        return False

pangram("The quick brown fox jumps over the lazy dog")
```

## 11. Escribe una funci贸n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab茅ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! お`

```python
def sort_alpha(string):
    words_list = []
    space = ''
    final_chain = ''
    for letter in string:
        if letter != ",":
            space = space + letter
        else:
            words_list.append(space)
            space = ''
    words_list.append(space)
    words_list.sort()

    for index, word in enumerate(words_list):
        final_chain += word
        if index < len(words_list) -1:
            final_chain += ','

    return final_chain
```

```python
# Para testear tu funci贸n 
test_alpha(sort_alpha)
```

## 12. Escribe una funci贸n para verificar si una contrase帽a dada es fuerte (al menos 8 caracteres, al menos una min煤scula, al menos una may煤scula, al menos un n煤mero y al menos un car谩cter especial). Deber铆a devolver True si es fuerte y False si no lo es.

```python
import re
def check_pass(password):
    pattern_minus = r'[a-z]'
    pattern_mayus = r'[A-Z]'
    pattern_num = r'[0-9]'
    pattern_special = r'[!@#$%\^&\*()\-_=\+[\]{};:\,<>\./?`~\\|]'
    
    if len(password) < 8:
        return False
    
    if not re.search(pattern_minus, password):
        return False
    
    if not re.search(pattern_mayus, password):
        return False
    
    if not re.search(pattern_num, password):
        return False
    
    if not re.search(pattern_special, password):
        return False
    
    return True
```

```python
# Para testear tu funci贸n
test_pass(check_pass)
```
